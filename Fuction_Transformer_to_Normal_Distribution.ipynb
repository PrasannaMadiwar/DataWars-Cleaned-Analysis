```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stast
```

```python
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title('PDF')
plt.subplot(122)
stast.probplot(X_train['Age'],dist='norm',plot=plt)
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
func_t = FunctionTransformer(func= np.log1p)
```

```python

func_t.fit(X_train['Age'])
```

```python
X_train__ = func_t.transform(X_train['Age'])
X_test__ = func_t.transform(X_test['Age'])
```

```python
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title('Befor')
plt.subplot(122)
sns.distplot(X_train__)
plt.title('After')
plt.show()
```

```python
stast.probplot(X_train__,dist='norm',plot=plt)
plt.show()
```
