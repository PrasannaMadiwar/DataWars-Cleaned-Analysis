```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.sample(4)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder

one = OneHotEncoder(drop='first',sparse_output=False, dtype=np.int32)
```

```python
one.fit(X_train[['Sex']])
```

```python
X_train_new = one.transform(X_train[['Sex']])
X_test_new = one.transform(X_test[['Sex']])
```

```python
X_train_new = np.hstack((X_train[['Pclass','Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare']].values,X_train_new))
```

```python
X_test_new = np.hstack((X_test[['Pclass','Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare']].values,X_test_new))
```

```python
feature_names = one.get_feature_names_out(['Sex'])

columns = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'] + list(feature_names)
```

```python
x_train = pd.DataFrame(X_train_new,columns=columns)
x_test = pd.DataFrame(X_test_new,columns=columns)
```
